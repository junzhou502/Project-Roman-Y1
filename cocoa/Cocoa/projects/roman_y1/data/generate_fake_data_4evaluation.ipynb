```python
import numpy as np
#---------------------You should modify and check this part--------------------------
nsrcs = 5
nlens = 5
nbins = 20
ggl_exclude = []

Keyword = None
#------------------------------------------------------------------------------------


ncombo = nsrcs*(nsrcs+1) + nlens*nsrcs + nlens - len(ggl_exclude)
ndv = nbins*ncombo
print('Number of tomographic correlation: ', ncombo)
print('Length of data vector: ', ndv)

# produce ones mask
out = np.zeros((ndv,2))
out[:, 0] = np.arange(0, ndv)
out[:, 1] = 1

if Keyword == None:
    np.savetxt(f'./roman_ones.mask', out ,fmt=["%d", "%.6e"])
    print(f'Saved roman_ones.mask')
else:
    np.savetxt(f'./roman_{Keyword}_ones.mask', out ,fmt=["%d", "%.6e"])
    print(f'Saved roman_{Keyword}_ones.mask')

# produce ones cov
out = np.zeros((int(ndv*(ndv+1)/2),10))
cnt = 0
for i in range(ndv):
    for j in range(i,ndv):
        out[cnt, 0] = i
        out[cnt, 1] = j
        if i==j:
            out[cnt, 8] = 1
            out[cnt, 9] = 0
        cnt+=1
print('Number of element:', cnt)
print('Shape of output file', out.shape)
if Keyword == None:
    np.savetxt(f'./cov_roman_eyes', out ,fmt=["%d", "%d", "%.6e", "%.6e", "%d", "%d", "%d", "%d", "%.6e", "%.6e"])
    print(f'Saved cov_roman_eyes')
else:
    np.savetxt(f'./cov_roman_{Keyword}_eyes', out ,fmt=["%d", "%d", "%.6e", "%.6e", "%d", "%d", "%d", "%d", "%.6e", "%.6e"])
    print(f'Saved cov_roman_{Keyword}_eyes')
```
